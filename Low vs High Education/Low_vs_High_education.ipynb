```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
higher_education = pd.read_csv('hoy_utdanning.csv')
low_education = pd.read_csv('lav_utdanning.csv')
```

```python
higher_education.head()
```

```python
low_education.head()
```

```python
# Choosing records for 2016
higher_education = higher_education.loc[higher_education['År'].isin([2016])]
low_education = low_education.loc[low_education['År'].isin([2016])]
```

```python
higher_education.head()
```

```python
# Percentage column is of dtype - object. In order to change the type to float we need to replace
# commas with dots.
higher_education['Variabel'] = higher_education['Variabel'].str.replace(',', '.')
low_education['Variabel'] = low_education['Variabel'].str.replace(',', '.')
```

```python
# Cleaning names of the districts, so that the same district is of the same format at all years.
higher_education['Navn'] = higher_education['Navn'].str.replace(r'\d+','').str.replace(' ','')
low_education['Navn'] = low_education['Navn'].str.replace(r'\d+','').str.replace(' ','')
```

```python
# Checking if the dataset looks right.
higher_education.head()
```

```python
# Checking if the dataset looks right.
low_education.head()
```

```python
# Choosing districts we want to explore
districts = ['Kannik', 'Bjergsted', 'Hundvågkrossen', 'Gausel', 'Jåttåvågen', 'Lervig', 'Stokka', 'Solborg',
            'Byhaugen', 'Buøy']
higher_education = higher_education.loc[higher_education['Navn'].isin(districts)] 
low_education = low_education.loc[low_education['Navn'].isin(districts)]
```

```python
# Converting columns with percentage values to float
convert_dict = {'Variabel': float}
higher_education = higher_education.astype(convert_dict)
low_education = low_education.astype(convert_dict)
```

```python
# Visualizing differences in share of population with lower vs higher education in chosen districts
fig, ax = plt.subplots(figsize=(12, 7))

ax = sns.barplot(x="Navn", y="Variabel", color='red', data=higher_education, label="High Education")
ax = sns.barplot(x="Navn", y="Variabel", color='blue', data=low_education, label="Low Education")

plt.title("Share of Districts Population with Higher vs Lower Education\n Ages 30 to 39", fontsize=18)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, fontsize='large')
ax.set_xlabel('Districts', fontsize=14) 
ax.set_ylabel('Percentage', fontsize=14)
ax.legend(title="Year")


figure = ax.get_figure()
#figure.savefig("higher_lower_ed.png")
```
